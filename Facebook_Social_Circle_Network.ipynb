# Loading in Dataset File - Into DF

```python
import pandas as pd
#before i got it to this stage, I personally extracted it from its .gz stage on the hyperlinked website above then added it to the directory here for this runtime
file_path = '/content/facebook_combined.txt'  #path to file in google colab directory

#load the file -> delimiter is a space between the source node column and target node for each row
df = pd.read_csv(file_path, sep=" ", header=None, names=["Source Node", "Target Node"])

#display the first few rows to check the data to make sure it correctly made it into a df
df.head()
```

```python
import networkx as nx

#create the graph from df
G = nx.from_pandas_edgelist(df, 'Source Node', 'Target Node')

#check basic information about the graph to see total nodes and edges we are working with
print(f"Graph info: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

#visualizing the graph of facebook social circles - essentially user interactions
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
nx.draw(G, node_size=30, with_labels=False, alpha=0.5)
plt.show()
```
